```python
for i in range(1960, 2016):
    fig, axis = plt.subplots(figsize=(12,6))
    dat = df.loc[df['year'] == i]

    sns.scatterplot(x='life_expectancy', y='fertility', hue='continent',
                    data=dat, size='population', sizes = (min_size, max_size), alpha=0.6, ax=axis)#, legend='brief') # size create point proportional to the population
    handles, labels = axis.get_legend_handles_labels()
    # Add country labels to each point
    for _, row in dat.iterrows():
        axis.text(row['life_expectancy'], 
                row['fertility'], 
                row['country'],
                fontsize= 5)

    axis.legend(handles=handles[:6], labels=labels[:6], loc=0) #handle specify how many  
    plt.title(f'life_exp vs fertility in {i}')
    plt.xlabel('Life expectancy')
    plt.ylabel('Fertility rate')
    
    plt.savefig(f'image/life_vs_fert_{i}.png')
    plt.close()
```
